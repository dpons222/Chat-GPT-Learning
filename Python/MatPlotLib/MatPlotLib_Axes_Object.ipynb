# Using Axes Object

```python
from matplotlib import pyplot as plt

drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales =  [91, 76, 56, 66, 52, 27]

plt.bar(range(len(drinks)), sales)

#create your ax object here
ax = plt.subplot()
ax.set_xticks(range(len(drinks)))
ax.set_xticklabels(drinks)



plt.show()
```

# Complex Axes

```python

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and axes
fig, axs = plt.subplots(2, 1, figsize=(8, 6))  # 2 rows, 1 column

# Plotting on the first axes
axs[0].plot(x, y1, color='blue')
axs[0].set_title('Sine Wave')
axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Y-axis')
axs[0].grid()

# Plotting on the second axes
axs[1].plot(x, y2, color='red')
axs[1].set_title('Cosine Wave')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Y-axis')
axs[1].grid()

plt.tight_layout()  # Adjusts the layout
plt.show()
```

# Complex Axes, Multiple Rows and Columns

```python

# Create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x / 5)
y5 = np.log(x + 1)
y6 = np.sqrt(x)

# Create a figure with 3 rows and 2 columns
fig, axs = plt.subplots(3, 2, figsize=(10, 12))  # 3 rows, 2 columns

# Plotting on each axes
axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('Sine Wave')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].grid()

axs[0, 1].plot(x, y2, color='red')
axs[0, 1].set_title('Cosine Wave')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')
axs[0, 1].grid()

axs[1, 0].plot(x, y3, color='green')
axs[1, 0].set_title('Tangent Wave')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')
axs[1, 0].grid()

axs[1, 1].plot(x, y4, color='orange')
axs[1, 1].set_title('Exponential Function')
axs[1, 1].set_xlabel('X-axis')
axs[1, 1].set_ylabel('Y-axis')
axs[1, 1].grid()

axs[2, 0].plot(x, y5, color='purple')
axs[2, 0].set_title('Logarithmic Function')
axs[2, 0].set_xlabel('X-axis')
axs[2, 0].set_ylabel('Y-axis')
axs[2, 0].grid()

axs[2, 1].plot(x, y6, color='cyan')
axs[2, 1].set_title('Square Root Function')
axs[2, 1].set_xlabel('X-axis')
axs[2, 1].set_ylabel('Y-axis')
axs[2, 1].grid()

plt.tight_layout()  # Adjusts the layout
plt.show()
```

## Yes, you can have multiple variables on the left side of an assignment in Python, and it's a feature called tuple unpacking. Here’s a brief explanation:

Tuple Unpacking
When you have a function (like plt.subplots()) that returns multiple values, you can assign those values to multiple variables in a single line.
The function returns a tuple (in this case, a figure and an array of axes), which can be unpacked directly into separate variables.
Example
In the case of fig, axs = plt.subplots(2, 1, figsize=(8, 6)), the plt.subplots() function returns two values:

**fig**: This is the figure object that represents the entire figure (the canvas for the plots).

**axs**: This is an array (or list) of axes objects where you can plot your data.
The function call effectively returns something like:

return (fig, axs)

So when you write fig, axs = ..., Python takes the returned tuple and assigns the first element to fig and the second element to axs.


# Understanding plt.subplots()
how do i know how many axs are there?

In the case of fig, axs = plt.subplots(2, 1, figsize=(8, 6)), you can determine the number of axes (axs) by looking at the arguments you passed to the plt.subplots() function. Here’s a breakdown of how it works:

The plt.subplots() function creates a grid of subplots and its parameters determine how many subplots (axes) are created. The syntax is:

fig, axs = plt.subplots(nrows, ncols, **kwargs)

- nrows: The number of rows of subplots you want.
- ncols: The number of columns of subplots you want.


For plt.subplots(2, 1, figsize=(8, 6)):

- nrows = 2: This means you want 2 rows of subplots.
- ncols = 1: This means you want 1 column of subplots.

So, the total number of axes created will be:

total axxes = nrows x ncols = 2 x 1 = 2




# Accessing the Axes



In this case, axs will be a NumPy array (or a list, depending on the configuration) that contains two axes objects. You can access each individual axis like this:

- axs[0]  # First axes (for the first subplot)
- axs[1]  # Second axes (for the second subplot)


## Accessing multiple Rows and column in Axes object

```python

# Create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x / 5)
y5 = np.log(x + 1)
y6 = np.sqrt(x)

# Create a figure with 3 rows and 2 columns
fig, axs = plt.subplots(3, 2, figsize=(10, 12))  # 3 rows, 2 columns

# Plotting on each axes
axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('Sine Wave')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].grid()

axs[0, 1].plot(x, y2, color='red')
axs[0, 1].set_title('Cosine Wave')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')
axs[0, 1].grid()

axs[1, 0].plot(x, y3, color='green')
axs[1, 0].set_title('Tangent Wave')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')
axs[1, 0].grid()

axs[1, 1].plot(x, y4, color='orange')
axs[1, 1].set_title('Exponential Function')
axs[1, 1].set_xlabel('X-axis')
axs[1, 1].set_ylabel('Y-axis')
axs[1, 1].grid()

axs[2, 0].plot(x, y5, color='purple')
axs[2, 0].set_title('Logarithmic Function')
axs[2, 0].set_xlabel('X-axis')
axs[2, 0].set_ylabel('Y-axis')
axs[2, 0].grid()

axs[2, 1].plot(x, y6, color='cyan')
axs[2, 1].set_title('Square Root Function')
axs[2, 1].set_xlabel('X-axis')
axs[2, 1].set_ylabel('Y-axis')
axs[2, 1].grid()

plt.tight_layout()  # Adjusts the layout
plt.show()
```

- axs[0, 0] for the first subplot (first row, first column).
- axs[0, 1] for the second subplot (first row, second column).
- axs[1, 0] for the third subplot (second row, first column).
- axs[1, 1] for the fourth subplot (second row, second column).
- axs[2, 0] for the fifth subplot (third row, first column).
- axs[2, 1] for the sixth subplot (third row, second column).
